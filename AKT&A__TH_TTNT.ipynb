```python
#-------THUẬT TOÁN AKT-------#
import copy
from heapq import heappush, heappop # Changed import statement

n = 3 # Giả sử Puzzle (n=3)
rows = [1, 0, -1, 0]
cols = [0, -1, 0, 1]

class priorityQueue: # Tạo mốt lớp hàng đợi
  def __init__(self):
    self.heap = [] # Corrected self.help to self.heap

  def push(self, key): # Chèn một khóa mới "key"
    heappush(self.heap, key)

  def pop(self): # Hàm xóa phần tử nhỏ nhất khỏi hàng đợi ưu tiên
    return heappop(self.heap)

  def empty(self):
    if not self.heap:
      return True
    else:
      return False

class nodes: # Cấu trúc của Node
  def __init__(self, parent, mats, empty_tile_posi, costs, levels):
    self.parent = parent                   # Lưu trữ nút cha của Node hiện tại và theo dõi đường đi khi giải pháp hiển thị
    self.mats = mats                       # Lưu trữ ma trận
    self.empty_tile_posi = empty_tile_posi # Lưu vị trí nơi mảnh gạch khoảng trống đã tồn tại trong ma trận
    self.costs = costs                     # Lưu lượng gạch đặt sai vị trí
    self.levels = levels                   # Lưu số lượt di chuyển cho đến nay

  def __lt__(self, nxt):#--Hàm tạo hàng đợi ưu tiên dựa trên giá trị chi phí của đối tượng--#
    return self.costs < nxt.costs

  @staticmethod # Added staticmethod decorator
  def calculateCosts(mats, final) -> int: #--Phương pháp tính số ô gạch bị sai vị trí--#
    count = 0
    for i in range(n):
      for j in range(n):
        if((mats[i][j]) and (mats[i][j] != final[i][j])):
          count +=1
    return count

  @staticmethod # Added staticmethod decorator
  def newNode(mats, empty_tile_posi, new_empty_tile_posi, levels, parent, final) -> 'nodes': # Added forward reference for type hint
    new_mats = copy.deepcopy(mats)          # Sao chép dữ liệu từ các ma trận gốc sang các ma trận hiện tại
    x1 = empty_tile_posi[0]                 # Di chuyển viên ô gạch 1 vị trí
    y1 = empty_tile_posi[1]
    x2 = new_empty_tile_posi[0]
    y2 = new_empty_tile_posi[1]
    new_mats[x1][y1], new_mats[x2][y2] = new_mats [x2][y2], new_mats[x1][y1]
    costs = nodes.calculateCosts(new_mats, final) # Corrected method call
    new_nodes = nodes(parent, new_mats, new_empty_tile_posi, costs, levels)
    return new_nodes

  @staticmethod # Added staticmethod decorator
  def printMatsrix(mats):#--Hàm in ma trận N x N--#
    for i in range(n):
      for j in range(n):
        print("%d"%(mats[i][j]), end = " ")
      print()

  @staticmethod # Added staticmethod decorator
  def isSafe(x, y):#--Hàm cho biết tọa độ (x,y) ma trận hợp lệ hay không--#
    return x >= 0 and x < n and y >= 0 and y < n

  @staticmethod # Added staticmethod decorator
  def printPath(root):#--Hàm in ra đường dẫn từ Node gốc dến Node cuối--#
    if root == None:
      return
    nodes.printPath(root.parent) # Corrected method call
    nodes.printMatsrix(root.mats) # Corrected method call
    print()

  @staticmethod # Added staticmethod decorator

  def solve(initial, empty_tile_posi, final):
    pq = priorityQueue()
    costs = nodes.calculateCosts(initial, final) # Corrected method call
    root = nodes (None, initial, empty_tile_posi, costs, 0)
    pq.push(root)
    while not pq.empty():
      minimum = pq.pop()
      if minimum.costs == 0:
        nodes.printPath(minimum) # Corrected method call
        return
      for i in range(n):
        new_tile_posi = [minimum.empty_tile_posi[0]+rows[i],
                         minimum.empty_tile_posi[1]+cols[i], ]
        if nodes.isSafe(new_tile_posi[0], new_tile_posi[1]): # Corrected method call
          child = nodes.newNode(minimum.mats, # Corrected method call
                           minimum.empty_tile_posi,
                           new_tile_posi,
                           minimum.levels + 1,
                           minimum, final,)
          pq.push(child)

initial = [ [1,2,3],
            [5,6,0],
            [7,8,4] ]

final = [ [1,2,3],
          [5,8,6],
          [0,7,4] ]

empty_tile_posi = [1, 2]
nodes.solve(initial, empty_tile_posi, final) # Corrected method call

print("-------------------------------------")

#-------THUẬT TOÁN A*-------#

from collections import deque

class Graph:
  def __init__(self, adjac_lis):
    self.adjac_lis = adjac_lis

  def get_neighbors(self, v):
    return self.adjac_lis[v]

  def h(self, n):
    H = {
        'A':1,
        'B':1,
        'C':1,
        'D':1
    }
    return H[n]

  def a_star_algorithm(self, start, stop):
    open_lst = set([start])
    closed_lst = set([])
    poo = {}
    poo [start] = 0
    par ={}
    par[start] = start
    while len(open_lst) > 0:
      n = None
      # Find the node with the lowest f = g + h in open_lst
      for v in open_lst:
        if n == None or poo[v] + self.h(v) < poo[n] + self.h(n):
          n = v

      if n == None:
        print('Path does not exist!')
        return None

      if n == stop:
        reconst_path = []
        while par[n] != n:
          reconst_path.append(n)
          n = par[n]
        reconst_path.append(start)
        reconst_path.reverse()
        print('Path found:{}'.format(reconst_path))
        return reconst_path

      # Remove the current node from the open_lst and add it to the closed_lst
      open_lst.remove(n)
      closed_lst.add(n)

      # Process neighbors of the current node
      for m, weight in self.get_neighbors(n):
        if m not in open_lst and m not in closed_lst:
          open_lst.add(m)
          par[m] = n
          poo[m] = poo[n] + weight
        else:
          if poo[m] > poo[n] + weight:
            poo[m] = poo[n] + weight
            par[m] = n
            if m in closed_lst:
              closed_lst.remove(m)
              open_lst.add(m)

    print('Path does not exist!')
    return None

adjac_lis = {
    'A':[('B',1), ('C',3), ('D',7)],
    'B':[('D',5)],
    'C':[('D',12)]
}
graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'D')
```
