```python
import networkx as nx
import matplotlib.pyplot as plt
# Đọc file
from google.colab import files
uploaded = files.upload()

def load_graph_from_content(file_content: str) -> list[list[int]]:
  """
  Loads a graph (adjacency matrix) from a string of file content.

  Args:
    file_content: A string containing the graph data, where each line
                  represents a row of the adjacency matrix and numbers
                  are space-separated.

  Returns:
    A list of lists of integers representing the adjacency matrix G.
  """
  G = []
  lines = file_content.strip().split('\n')
  for line in lines:
    row = [int(x) for x in line.split()]
    G.append(row)
  return G
print("Function 'load_graph_from_content' defined.")
if 'uploaded' in locals() and uploaded:
  first_uploaded_filename = list(uploaded.keys())[0]
  file_content_string = uploaded[first_uploaded_filename].decode('utf-8')
  # Load the graph using the new function
  loaded_G = load_graph_from_content(file_content_string)
  print(f"Loaded graph G from '{first_uploaded_filename}':")
  for row in loaded_G:
    print(row)
else:
  print("No file content found in 'uploaded'. Please run the file upload cell first.")

node = "ABCDEF"
t_ ={}
for i in range(len(G)):
  t_[node[i]] = i

degree =[]
for i in range(len(G)):
  degree.append(sum(G[i]))

colorDict = {}
for i in range(len(G)):
  colorDict[node[i]] = ["Blue","Red","Yellow","Green"]

sortedNode = []
indeks = []

for i in range(len(degree)):
  _max = 0
  j = 0
  for j in range(len(degree)):
    if j not in indeks:
      if degree[j] > _max:
        _max = degree[j]
        idx = j
  indeks.append(idx)
  sortedNode.append(node[idx])

theSolution = {}
for n in sortedNode:
  setTheColor = colorDict[n]
  if not setTheColor:
    print(f"Warning: No colors available for node {n}. Assigning 'Uncolored'.")
    theSolution[n] = "Uncolored"
    continue

  theSolution[n] = setTheColor[0]
  chosen_color = theSolution[n]

  adjacentNode = G[t_[n]]
  for j in range(len(adjacentNode)):
    if adjacentNode[j] == 1 and (chosen_color in colorDict[node[j]]):
      colorDict[node[j]].remove(chosen_color)

for t,w in sorted(theSolution.items()):
  print("Đỉnh ",t," = ",w)

```

## Create Graph and Map Colors

### Subtask:
Create a NetworkX graph object from the adjacency matrix `G` and prepare a list of colors for each node based on `theSolution`.


**Reasoning**:
The subtask requires creating a NetworkX graph from the adjacency matrix `G` and then generating a list of node colors based on the `theSolution` dictionary. This code block will initialize a directed graph, add edges based on the `G` matrix, and collect the corresponding colors for each node in the order they appear in the graph.



```python
graph = nx.DiGraph()

num_nodes = len(G)
for i in range(num_nodes):
    for j in range(num_nodes):
        if G[i][j] == 1:
            graph.add_edge(node[i], node[j])

node_colors = []
# Ensure the order of colors matches the order of nodes in the graph
for n_name in graph.nodes():
    node_colors.append(theSolution[n_name])

print("NetworkX graph created with", graph.number_of_nodes(), "nodes and", graph.number_of_edges(), "edges.")
print("Node colors list prepared.")
```

## Visualize Graph

### Subtask:
Draw the NetworkX graph with assigned node colors and labels.


**Reasoning**:
The subtask is to visualize the graph with assigned node colors and labels. This code block will draw the graph using NetworkX and Matplotlib, applying the node colors and labels as specified in the instructions.



```python
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(graph)
nx.draw_networkx(graph, pos, with_labels=True, node_color=node_colors, node_size=1500, font_size=10, font_color="black", edge_color="gray", alpha=0.7)
plt.axis('off')
plt.title("Graph G with Node Coloring")
plt.show()
```
